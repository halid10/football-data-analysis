```python
# Import the libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# Load Dataset

file = "EPL_20_21.csv"
epl_df = pd.read_csv(file)
epl_df.head()
```

```python
epl_df.info()
```

```python
epl_df.describe()
#epl_df.describe(include="all")
```

```python
# check total number of null values for each column

epl_df.isna().sum()
```

```python
# create 2 new columns : MinsPerMatch and GoalsPerMatch

epl_df['MinsPerMatch'] = (epl_df['Mins']/epl_df['Matches']).astype(int)
epl_df['GoalsPerMatch'] = (epl_df['Goals']/epl_df['Matches']).astype(float)
epl_df.head()
```

```python
# Total Goals, penalty and penalty attempts in the season

total_goals = epl_df['Goals'].sum()
total_penaltyGoals = epl_df['Penalty_Goals'].sum()
total_penaltyAttempts = epl_df['Penalty_Attempted'].sum()

print("The EPL season 20/21 has registered a total of {} goals, including {} penalties from {} attempts." .format(total_goals, total_penaltyGoals, total_penaltyAttempts))
```

```python
# Pie chart for penalties missed vs scored

plt.figure(figsize=(13, 6))
pen_scored = total_penaltyGoals
pen_missed = total_penaltyAttempts-total_penaltyGoals
data = [pen_scored, pen_missed]
label = ['Penalties scored', 'Penalties missed']
color = sns.color_palette('Set2')

plt.pie(data, labels=label, colors=color, autopct='%.0f%%')
plt.show
```

```python
# Most players from which countries (top 10 nationalities)

#nationality = epl_df.groupby('Nationality').size().sort_values(ascending=False)
nationality = epl_df['Nationality'].value_counts()
nationality.head(10).plot(kind='bar', figsize=(12, 6), color=sns.color_palette('magma'))
```

### Unsurprisingly, England is the most represented country 

```python
# Clubs with maximum players in their squad

epl_df['Club'].value_counts().nlargest(5).plot(kind='bar', figsize=(12,6), color=sns.color_palette('viridis'))
```

```python
# Clubs with least players in their squad

epl_df['Club'].value_counts().nsmallest(5).plot(kind='bar', figsize=(12,6), color=sns.color_palette('viridis'))
```

```python
# Players based on age group

under20 = epl_df[epl_df['Age']<=20]
age20_25 = epl_df[(epl_df['Age']>20) & (epl_df['Age']<=25)]
age25_30 = epl_df[(epl_df['Age']>25) & (epl_df['Age']<=30)]
above30 = epl_df[epl_df['Age']>30]
```

```python
x = np.array([under20['Name'].count(), age20_25['Name'].count(), age25_30['Name'].count(), above30['Name'].count()])
mylabels = ["<=20", ">20 & <=25", ">25 & <=30", ">30"]
plt.title('Total Players with Age', fontsize=20)
plt.pie(x, labels=mylabels, autopct='%.1f%%')
plt.show()
```

### 60% of players are aged between 21 and 30

```python
#Total under 20 players in each club

players_under_20 = epl_df[epl_df['Age']<20]
players_under_20['Club'].value_counts().plot(kind='bar', color=sns.color_palette("cubehelix"))
```

```python
# Average age of players in each club

plt.figure(figsize=(12, 6))
sns.boxplot(x = 'Club', y='Age', data = epl_df)
plt.xticks(rotation=90)
```

```python
data = epl_df.groupby('Club')['Age'].mean()
data.sort_values(ascending=False)
```

### Crystal Palace is the club with the highest average age players, i.e most of their players are older 

```python
# Total assists from club

assists_by_clubs = pd.DataFrame(epl_df.groupby('Club', as_index=False)['Assists'].sum())
sns.set_theme(style="whitegrid", color_codes=True)
ax = sns.barplot(x='Club', y='Assists', data=assists_by_clubs.sort_values(by='Assists'), palette='Set2')
ax.set_xlabel('Club', fontsize=20)
ax.set_ylabel('Assists', fontsize=20)
plt.xticks(rotation=75)
plt.rcParams["figure.figsize"]=(20,8)
plt.title('Plot of Clubs vs Total Assists', fontsize=20)
```

```python
# Top 10 assists

top_10_assists = epl_df[['Name', 'Club', 'Assists', 'Matches']].nlargest(n=10, columns='Assists')
top_10_assists
```

### Haary Kane (14) had the most assists

```python
# Total goals from club

goals_by_clubs = pd.DataFrame(epl_df.groupby('Club', as_index=False)['Goals'].sum())
sns.set_theme(style="whitegrid", color_codes=True)
ax = sns.barplot(x='Club', y='Goals', data=goals_by_clubs.sort_values(by='Goals'), palette='Set2')
ax.set_xlabel('Club', fontsize=20)
ax.set_ylabel('Goals', fontsize=20)
plt.xticks(rotation=75)
plt.rcParams["figure.figsize"]=(20,8)
plt.title('Plot of Clubs vs Total Goals', fontsize=20)
```

### Manchester City had the highest goals at 82 scored followed by Manchester United (70) and Tottenham Hotspur (66). The lowest scoring team was Sheffield United (19), followed by Fulham (26) then Burnley (32).

```python
# Pie chart - Goals with assist and without assist

plt.figure(figsize=(14,7))
assists = epl_df['Assists'].sum()
data = [total_goals-assists, assists]
labels = ['Goals without assists', 'Goals with assists']
color = sns.color_palette('Set1')
plt.pie(data, labels=labels, colors=color, autopct='%.0f%%')
plt.show()
```

### 31% of goals were scored without assists while 69% with assists 

```python
# Top 5 players with most cards

epl_df['Total_Cards'] = epl_df['Yellow_Cards'] + epl_df['Red_Cards']
epl_cards = epl_df.sort_values(by='Total_Cards', ascending=False)[:5]
plt.figure(figsize=(20,6))
plt.title("Most aggressive players", fontsize=20)
c=sns.barplot(x=epl_cards['Name'], y=epl_cards['Total_Cards'], color='red')
plt.xlabel('Players', fontsize=20)
plt.ylabel('Total Cards', fontsize=20)
c.set_xticklabels(c.get_xticklabels(), rotation=45)
c
```

### John McGinn (12), Douglas Luiz (11), Conor Gallagher (11), Harry Maguire (11), and Kalvin Phillips (10) are the most aggressive players in the EPL 2020-21 as they have the highest number of red and yellow cards.
